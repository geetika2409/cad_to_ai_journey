```python
## Arithematic operator

driver1 = 72
driver2 = 63     ## driver1 and driver2 are from same team
driver3 = 49

total_races = 24

races_left = 21

team_score = driver1 + driver2
print("Total team score: ", team_score)

gap_between_drivers = driver1 - driver2
print("Gap between drivers: ", gap_between_drivers)

points_left = races_left * 25
print("Points left in race to score: ", points_left)

avg_point = driver3 / (total_races-races_left)
print("Average points scored by driver 3 in each race: ", avg_point)

## floor division
avg_point = driver3 // (total_races-races_left)                        
print("Average points scored by driver 3 in each race: ", avg_point)

## modulus
driver_non_win_runs = driver1 % 25
print("Points scored by driver by lost matches: ", driver_non_win_runs)

## exponential
squarred_points_gap = (driver1**2) - (driver2**2)
print("Gap between drivers: ", squarred_points_gap)
```

```python
## Comparison Operator

driver1 = 72
driver2 = 63     ## driver1 and driver2 are from same team
driver3 = 49

print(driver1 == driver2)

name1 = "Russell"
name2 = "russell"
print(name1 == name2)

## Not equal to
print(name1 != name2)

print(driver1 > driver2)

print(driver1 < driver2)
```

```python
## Logical Operator

driver1_winner = True
driver2_winner = False

print(driver1_winner and driver2_winner)

print(driver1_winner or driver2_winner)

print(not driver2_winner)
```
