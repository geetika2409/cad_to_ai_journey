```python
## range

range(5)

print(range(5))
```

```python
## range with for loop

for i in range(5):
    print(i)
```

```python
## range with start and stop

for i in range(1, 5):
    print(i)
```

```python
## range with start and stop and step

for i in range(1, 10, 3):
    print(i)
```

```python
## range with step reverse

for i in range(10, 0, -3):
    print(i)
```

```python
## string with for loop

string = "Max Verstappen"

for char in string:
    print(char)
```
