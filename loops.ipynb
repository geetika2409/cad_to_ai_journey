```python
## range

range(5)

print(range(5))
```

```python
## range with for loop

for i in range(5):
    print(i)
```

```python
## range with start and stop

for i in range(1, 5):
    print(i)
```

```python
## range with start and stop and step

for i in range(1, 10, 3):
    print(i)
```

```python
## range with step reverse

for i in range(10, 0, -3):
    print(i)
```

```python
## string with for loop

string = "Max Verstappen"

for char in string:
    print(char)
```

```python
## while loop

cnt = 1

while (cnt < 5):
    print(cnt)
    cnt = cnt + 1
```

```python
## Loop control statements

## break

for i in range(10):
    if i==4:
        break
    print(i)
```

```python
## continue

for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## pass statement

for i in range(5):
    if i==3:
        print("The number is: ", i)
        pass
    print(i)
```

```python
## Nested loops

for i in range(5):
    for j in range(5):
        print(f"i : {i} and  j : {j}")
```

```python
## Prime numbers from 1 to 50

for i in range(1, 51):
    if i>1:
        for j in range(2, i):
            if i%j == 0:
                break
        else:
                print(i)
```
