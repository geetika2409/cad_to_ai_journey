```python
## Basic syntax rules in Python

## Case sensitivity - Python is case-sensitive

name = "Krish"
Name = "Naik"

print(name)
print(Name)
```

```python
## indentation - python uses indentation to define block of codes

age = 26
if age>30:
    print(age)   ## this line won't be printed as considered part of the if block

print(age)
```

```python
## line continuation

total = 1+2+3+4+5+6+7+\
4+5+6

print(total)
```

```python
## multiple statements in single line

x = 5; y = 6; z = x+y
print(z)
```

```python
## Semantics in python

## Variable assignment

age = 32
name = "Krish"

print(type(age))
```

```python
print(type(name))
```

```python
## type inference

variable = 10
print(type(variable))

variable = "Krish"
print(type(variable))
```

```python
## indentation example

if True:
    print("Tesla is the best")
    if False:
        print("This is not printable")
    print("This will print")
print("This is out of the print block")
```
