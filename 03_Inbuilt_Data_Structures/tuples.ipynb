```python
## Creating a Tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## List_to_Tuple_conversion_viceversa

lst = [1, 2, 3, 4, 5, 6, 7]
tup = tuple(lst)
print(f" Tuple: {tup}")

lst1 = list(tup)
print(f" List: {lst1}")

```

```python
## Mixed tuple

mxd_tup = (1, 3.14, "Mercedes", True)
print(mxd_tup)
```

```python
## Accessing elements of Tuples

print(tup)

print(tup[2])

print(tup[: 4])

print(tup[: : -1])
```
