```python
## Creating a Tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## List_to_Tuple_conversion_viceversa

lst = [1, 2, 3, 4, 5, 6, 7]
tup = tuple(lst)
print(f" Tuple: {tup}")

lst1 = list(tup)
print(f" List: {lst1}")

```

```python
## Mixed tuple

mxd_tup = (1, 3.14, "Mercedes", True)
print(mxd_tup)
```

```python
## Accessing elements of Tuples

print(tup)

print(tup[2])

print(tup[: 4])

print(tup[: : -1])
```

```python
## Tuple Operations

## Concatenation

tup = (1, 2, 3, 4, 5, 6, 7)

concatenated_list = tup + mxd_tup
print(f" Concatenated Tuple: ", concatenated_list)
```

```python
## Repeation

print(mxd_tup * 3)
```

```python
## Tuple Methods

f1_set = (54, 56, 54, "Kimi Antonelli", "Max Verstappen", 72)

print(f"Count of 54 in set is: {f1_set.count(54)}")
print(f"Index of 54 in set is: {f1_set.count(54)}")
```

```python
## Packing and Unpacking

## Packing

packed_tuple = 1, "Hello", 3.14
print(packed_tuple)
```

```python
## Unpacking Tuple

a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking tuple with *

first, *middle, last = (1, 2, 3, 4, 5, 6)
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples

tup_nested = ([1, 23, 4], (3, 4, 5), ("Isha", "Rockstar"))
print(f" Nested Tuple: {tup_nested}")
```

```python
## Iterating over nested tuple

for subtup in tup_nested:
    for item in subtup:
        print(item)
```
