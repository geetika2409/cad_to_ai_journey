```python
## defining a list

lst = []
print(type(lst))
```

```python
## mixed list printing

mixed_list = [1, "Isha", 2.13, True]
print(mixed_list)
```

```python
## Accessing list elements 

lst = ["Mercedes", "Aston martin", "Red Bull", "Haas", "Red Bull Racing"]

print(lst[0])
print(lst[3])
print(lst[-1])
```

```python
print(lst[1:2])
print(lst[1:])
print(lst[-1:])
```

```python
## Modifying the list

lst = ["Mercedes", "Aston martin", "Red Bull", "Haas", "Red Bull Racing"]

lst[1] = "Scuderia Ferrari"
print(lst)
```

```python
## list methods

## apend

lst.append("Mc Laren")
print(lst)
```

```python
## insert

lst.insert(1, "Racing Bulls")
print(lst)
```

```python
## remove

lst.remove("Racing Bulls")
print(lst)
```

```python
## pop

popped_element = lst.pop()
print(popped_element)
print("List :", lst)
```

```python
## index

indx = lst.index("Red Bull")
print(indx)
```

```python
## count

lst.remove("Red Bull")

print(lst)

print(lst.count("Red Bull"))
```

```python
## sort

lst.sort()
print("List in ascending order: ", lst)
```

```python
## reverse

lst.reverse()
print("List : ", lst)
```

```python
## clear

lst.clear()
print("List: ", lst)
```

```python
## Slicing Lists

drivers = ["Max Verstappen", "Kimi Antonelli", "Oscar Piastri", "Lando Norris", "Charles Lecrec", "Lewis Hamilton"]

print(drivers[2:5])
print(drivers[:5])
print(drivers[5:])
print(drivers[::5])
print(drivers[::-1])
```
