```python
## defining a list

lst = []
print(type(lst))
```

```python
## mixed list printing

mixed_list = [1, "Isha", 2.13, True]
print(mixed_list)
```

```python
## Accessing list elements 

lst = ["Mercedes", "Aston martin", "Red Bull", "Haas", "Red Bull Racing"]

print(lst[0])
print(lst[3])
print(lst[-1])
```

```python
print(lst[1:2])
print(lst[1:])
print(lst[-1:])
```

```python
## Modifying the list

lst = ["Mercedes", "Aston martin", "Red Bull", "Haas", "Red Bull Racing"]

lst[1] = "Scuderia Ferrari"
print(lst)
```

```python
## list methods

## apend

lst.append("Mc Laren")
print(lst)
```

```python
## insert

lst.insert(1, "Racing Bulls")
print(lst)
```

```python
## remove

lst.remove("Racing Bulls")
print(lst)
```

```python
## pop

popped_element = lst.pop()
print(popped_element)
print("List :", lst)
```

```python
## index

indx = lst.index("Red Bull")
print(indx)
```

```python
## count

lst.remove("Red Bull")

print(lst)

print(lst.count("Red Bull"))
```

```python
## sort

lst.sort()
print("List in ascending order: ", lst)
```

```python
## reverse

lst.reverse()
print("List : ", lst)
```

```python
## clear

lst.clear()
print("List: ", lst)
```

```python
## Slicing Lists

drivers = ["Max Verstappen", "Kimi Antonelli", "Oscar Piastri", "Lando Norris", "Charles Lecrec", "Lewis Hamilton"]

print(drivers[2:5])
print(drivers[:5])
print(drivers[5:])
print(drivers[::5])                   ## taking elements from starting maintaining step size of 5
print(drivers[::-1])                  ## printing elements from the reverse
```

```python
## iterating over the list

for driver in drivers:
    print(driver)
```

```python
## iterating with index

for indx, driver in enumerate(drivers):      ## enumerate function return index and value at specified index
    print(indx, driver)
```

```python
## List comprehension

lst = []

for i in range(10):
    lst.append(i**2)

print(lst)


## Applying list comprehension

lst = [i**2 for i in range(10)]
print(lst)
```

```python
## List comprehension with condition

lst = []
for i in range(10):
    if i%2 == 0:
        lst.append(i**2)

print(lst)

## List applying list comprehension

lst = [i**2 for i in range(10) if i%2 == 0]
print(lst)
```

```python
## Nested list comprehension

lst1 = [1, 2, 3, 4]
lst2 = ['a','b', 'c', 'd', 'e']

pair = [(i, j) for i in lst1 for j in lst2]
print(pair)
```

```python
## list comprehension with function calls

drivers = ["Max Verstappen", "Kimi Antonelli", "Oscar Piastri", "Lando Norris", "Charles Lecrec", "Lewis Hamilton"]

lengths = [len(word) for word in drivers]
print(lengths)
```
