### Assignment 6: List Methods

Create a list of random numbers and sort it in ascending and descending order. Remove the duplicates from the list and print the modified list.

```python
import random

random_num = [random.randint(1, 20) for _ in range(15)]
print(f"Original list: ", random_num)

ascending_list = sorted(random_num)
print("Ascending List: ", ascending_list)

descending_list = sorted(random_num, reverse = True)
print("Descending List: ", descending_list)

unique_list = set(random_num)
print("Modified List: ", unique_list)
```

### Assignment 8: List of Dictionaries

Create a list of dictionaries where each dictionary represents a student with keys 'name' and 'score'. Sort the list of dictionaries by the 'score' in descending order and print the sorted list.

```python
drivers = [
    {'name' : "Kimi Antonelli", 'points' : 24},
    {'name' : "Lando Norris", 'points' : 72},
    {'name' : "Max Verstappen", 'points' : 63},
    {'name' : "Oscar Piastri", 'points' : 80},
    {'name' : "Lewis Hamilton", 'points' : 98}
]

sorted_drivers = sorted(drivers, key = lambda x: x['points'], reverse = True)

print("Sorted list of drivers: - ")

for driver in sorted_drivers:
    print(driver)
```

### Assignment 9: Matrix Transposition

Write a function that takes a 3x3 matrix (nested list) as input and returns its transpose. Print the original and transposed matrices.

```python
def transpose_mat(mat):
    transposed = [[mat[i][j] for j in range(len(mat))] for i in range(len(mat[0]))]
    return transposed

mat = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Original Matrix: ")

for row in mat:
    print(row)

transposed = transpose_mat(mat)

print("Transposed Matrix: ")

for row in transposed:
  print(row)
```

### Assignment 10: Flattening a Nested List

Write a function that takes a nested list and flattens it into a single list. Print the original and flattened lists.

```python
def flatten_lst(lst):
    flat_lst = [item for sublist in lst for item in sublist]
    return flat_lst

lst = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print("Original List: ")
print(lst)

print("Flattened List: ")
print(flatten_lst(lst))
```

### Assignment 15: List Intersection

Write a function that takes two lists and returns a new list containing only the elements that are present in both lists. Print the intersected list.

```python
def list_intersection(lst1, lst2):
    lst = [x for x in lst1 if x in lst2]
    return lst

lst1 = [1, 2, 3, 4, 5, "a", "b", "c"]
lst2 = [1, 3, 3, 4, 5, "e", "g", "c"]

intersected_list = list_intersection(lst1, lst2)

print(f" Intersected List: {intersected_list}")
```
