```python
## if statement

driver1_points = 72
driver2_points = 63

if (driver1_points > driver2_points):
    print("Driver1 is the winner")
if (driver1_points < driver2_points):
    print("Driver2 is the winner")
```

```python
## else statement

driver1_points = 72
driver2_points = 63

if (driver1_points > driver2_points):
    print("Driver1 is the winner")
else:
    print("Driver2 is the winner")
```

```python
## elif

driver1_points = 72
driver2_points = 63
driver3_points = 49

if (driver1_points > driver2_points and driver2_points > driver3_points):
    print("Driver1 is the winner")
elif (driver2_points > driver1_points and driver2_points > driver3_points):
    print("Driver2 is the winner")
else:
    print("driver3 is the winner")
```

```python
## nested conditional statement

points = 72
wins = 2

if (points > 50 and wins >= 2):
    print("Title Contender")

elif (points > 50 and wins >= 1):
    print("Strong Challenger")

elif (points > 25 and wins >= 1):
    print("Occasional Scorer")

elif (points > 0):
    print("Backmarker")

else:
    print("No Points Yet")
```
